```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from word2number import w2n
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("hiring.csv")
df.head()
```

```python
df.describe()
```

```python
m = df["test_score(out of 10)"].mean()
df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(m)
```

```python
df.experience.fillna("zero",inplace=True)
```

```python
df["experience"] = df["experience"].apply((w2n.word_to_num))
                                   
```

```python
X= df.drop(columns=["salary($)"])
X
```

```python
y=df["salary($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
predicted =lr.predict(X)
predicted
```

```python
df["pridicted"]=predicted
df
```

```python
X.head(1)
```

```python
exp= float(input("Experience"))
t_score= float(input("Test Score"))
i_score=float(input("Interview Score"))
user_pred= lr.predict([[exp,t_score,i_score]])
print("Predicted Salary $: ",user_pred[0].round(decimals=2))
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.index,y,label="Actual Salary",marker ="o")
plt.plot(df.index,predicted,label="Pridicted Salary",marker="X")
plt.scatter(len(df),user_pred,color="red",label="your Predicted Salary",s=50,marker="*")
plt.xlabel("Data point Index")
plt.ylabel("Salary (&)")
plt.title("Actual vs Predicted Salary")
plt.legend()
plt.grid(True)
plt.show()
```
